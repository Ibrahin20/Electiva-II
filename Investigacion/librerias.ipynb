PyTorch (Aprendizaje profundo y redes neuronales)
Similar a TensorFlow, pero con una estructura más flexible para investigación en IA.

```python
# 1. Importar librerías
import torch
import torch.nn as nn
from torch.optim import SGD

# 2. Generar datos de ejemplo (CPU/GPU)
torch.manual_seed(42)
X = torch.linspace(0, 5, 100).reshape(-1, 1)  # Característica
y = 2.5 * X + 1.2 + torch.randn(X.size())      # Target con ruido

# 3. Definir modelo
class RegresionLineal(nn.Module):
    def __init__(self):
        super().__init__()
        self.capa = nn.Linear(1, 1)  # 1 entrada, 1 salida
        
    def forward(self, x):
        return self.capa(x)

modelo = RegresionLineal()

# 4. Configurar entrenamiento
criterio = nn.MSELoss()               # Función de pérdida
optimizador = SGD(modelo.parameters(), lr=0.01)  # Optimizador

# 5. Bucle de entrenamiento
for epoch in range(200):
    # Forward pass
    predicciones = modelo(X)
    perdida = criterio(predicciones, y)
    
    # Backward pass
    optimizador.zero_grad()
    perdida.backward()
    optimizador.step()
    
    if epoch % 50 == 0:
        print(f'Época {epoch}, Pérdida: {perdida.item():.4f}')

# 6. Hacer predicción
with torch.no_grad():
    nuevo_X = torch.tensor([[3.0]])
    prediccion = modelo(nuevo_X)
    print(f'\nPredicción para valor 3: {prediccion.item():.2f}')
```
